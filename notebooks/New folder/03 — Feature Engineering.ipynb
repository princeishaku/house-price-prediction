Load Clean Dataset

```python
train = pd.read_csv("../data/processed/train_clean.csv")

train.head()
```

Section 1 — Distribution of SalePrice

This is the most important variable because it's what we'll predict.

```python
plt.figure(figsize=(10,6))

sns.histplot(
    train["SalePrice"],
    bins=40,
    kde=True
)

plt.title("Distribution of SalePrice")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")

plt.show()
```

Questions to answer in your notebook

Section 2 — Correlation Heatmap

First, calculate correlations for numeric features.

```python
corr = train.corr(numeric_only=True)
```

```python
plt.figure(figsize=(18,14))

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Heatmap")

plt.show()
```

This gives a global view of relationships between numeric variables.

Section 3 — Top 20 Features Correlated with SalePrice

Instead of the full heatmap, focus on the target variable.

```python
top_corr = (
    corr["SalePrice"]
    .sort_values(ascending=False)
    .head(20)
)

top_corr
```

Visualize:

```python
plt.figure(figsize=(10,8))

sns.barplot(
    x=top_corr.values,
    y=top_corr.index
)

plt.title("Top Features Correlated with SalePrice")

plt.xlabel("Correlation")

plt.show()
```

Section 6 — Scatter Plot

Relationship between living area and price.

Section 7 — Boxplot by Overall Quality

Section 8 — Correlation Matrix (Top 15)

```python
top15 = corr["SalePrice"].abs().sort_values(ascending=False).head(15).index

plt.figure(figsize=(12,10))

sns.heatmap(
    train[top15].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Top 15 Correlated Features")

plt.show()
```

This helps identify multicollinearity among important predictors.

Create features and target

```python
from sklearn.model_selection import train_test_split

# Target variable
y = train["SalePrice"]

# Features
X = train.drop(columns=["SalePrice"])

print(X.shape)
print(y.shape)
```

Split the data

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training set:", X_train.shape)
print("Validation set:", X_valid.shape)
```

Create the output folder

```python
from pathlib import Path

Path("../data/processed").mkdir(
    parents=True,
    exist_ok=True
)
```

```python
X_train.to_csv(
    "../data/processed/X_train.csv",
    index=False
)

X_valid.to_csv(
    "../data/processed/X_valid.csv",
    index=False
)

y_train.to_csv(
    "../data/processed/y_train.csv",
    index=False
)

y_valid.to_csv(
    "../data/processed/y_valid.csv",
    index=False
)

print("All processed datasets saved successfully!")
```
